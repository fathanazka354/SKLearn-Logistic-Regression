```python
import pandas as pd

# reading datasets and using it to dataframe
df = pd.read_csv('./sample_data/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
# check whether kind of empty data or no
df.info()
```

```python
# drop column that not needed
data = df.drop(columns=['User ID'])

# running proccess of encoding one-hot with pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
# slice between attribute and label
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']
X = data[predictions]
y = data['Purchased']
```

```python
# doing normalisation with data that my own
from sklearn.preprocessing import StandardScaler as ss
scaler = ss()
scaler.fit(X)
```

```python
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split as tts

# divide data to train and test for each of attribute and label
X_train, X_test, y_train, y_test = tts(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# training model using fit function
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# test accuration of model
model.score(X_test, y_test)
```
